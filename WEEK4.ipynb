##문제 1

함수 이름: function1

설명

이 함수는 개수가 정해지지 않은 위치 인자를 입력받는다  
모든 인자는 정수이다  

조건

60점 이상인 값들만 골라 합을 반환하시오  
조건을 만족하는 값이 없으면 0을 반환하시오  

예시

function1(10, 60, 70, 50) -> 130  
function1(60, 70, 80) -> 210  
function1(10, 20, 30) -> 0  

### 힌트
- `*args`는 여러 개의 위치 인자를 받을 때 사용합니다.
- 반복문(`for`)으로 모든 값을 하나씩 확인할 수 있습니다.
- 조건문(`if`)을 사용해서 60 이상인 값만 골라보세요.
- 합을 저장할 변수를 먼저 만들어두면 편합니다.

```python
def function1(*args):
  result = 0
  for i in args:
    if i >= 60 :
      result += i
  return result
```

##문제 2

함수 이름: function2

설명

이 함수는 하나의 위치 인자를 입력받는다  
문자열은 숫자와 알파벳으로 구성되어 있다  

조건

숫자는 모두 제거하시오  
알파벳만 남기고 모두 소문자로 변환하여 반환하시오  

예시

function2("A123b45c") -> "abc"  
function2("123Hello123") -> "hello"  
function2("2025Happy") -> "happy"  

### 힌트
- 문자열은 반복문으로 한 글자씩 확인할 수 있습니다.
- 알파벳인지 확인하려면 `isalpha()`를 사용할 수 있습니다.
- 숫자인지 확인하려면 `isdigit()`을 사용할 수 있습니다.
- 조건을 만족하는 문자만 모아서 새로운 문자열을 만들 수 있습니다.
- 대소문자를 바꾸려면 `lower()`를 사용할 수 있습니다.

```python
def function2(args):
  alpha = ""
  for i in args:
    if i.isalpha() == True:
      alpha += i
  alpha = alpha.lower()
  return alpha
```

##문제 3

함수 이름: function3

설명

이 함수는 개수가 정해지지 않은 키워드 인자를 입력받는다  
모든 값은 정수이다  

조건

key의 길이가 2 이상인 경우만 선택한다  
해당 value들을 모두 더하여 반환하시오  
조건을 만족하는 값이 없으면 0을 반환하시오  

예시

function3(a=5, bb=10, c=3) -> 10  
function3(aa=5, bb=10) -> 15  
function3(a=1, b=2) -> 0  

### 힌트
- `**kwargs`는 key와 value 형태의 데이터를 받습니다.
- `for key, value in kwargs.items()` 형태로 순회할 수 있습니다.
- key의 길이는 `len(key)`로 확인할 수 있습니다.
- 조건을 만족하는 value만 더해보세요.

```python
def function3(**kwargs):
  result = 0
  for key, value in kwargs.items():
    if len(key) >= 2:
      result += kwargs[key]
  return result
```

##문제 4

함수 이름: function4

설명

이 함수는 하나의 위치 인자를 입력받는다  
정수로 이루어진 리스트이다  

조건

짝수만 선택하여 더하시오  
음수는 모두 제외하시오  
조건을 만족하는 값이 없으면 0을 반환하시오  

예시

function4([1, 2, 3, 4]) -> 6  
function4([-2, 2, 4]) -> 6  
function4([1, 3, 5]) -> 0  

### 힌트
- 리스트는 반복문(`for`)으로 하나씩 확인할 수 있습니다.
- 짝수인지 확인하려면 `% 2 == 0` 조건을 사용할 수 있습니다.
- 음수를 제외하려면 0 이상인지 먼저 확인해보세요.
- 조건을 만족하는 값만 더하도록 `if`문을 활용하세요.

```python
def function4(args):
  result = 0
  for i in args:
    if i > 0 :
      if i % 2 == 0 :
        result += i
  return result
```

##문제 5

함수 이름: function5

설명

이 함수는 개수가 정해지지 않은 위치 인자와 키워드 인자를 입력받는다  
모든 값은 정수이다  

조건

위치 인자(args)에서는 짝수만 골라 더한다  
키워드 인자(kwargs)에서는 key가 'a'로 시작하는 값만 더한다  
두 결과를 합하여 반환하시오  

예시

function5(1, 2, 3, 4, a=5, b=10) -> 11  
function5(2, 4, 6, aa=3, ab=5) -> 20  
function5(1, 3, 5) -> 0  

### 힌트
- `*args`와 `**kwargs`를 동시에 사용할 수 있습니다.
- `args`는 여러 개의 위치 인자이므로 반복문으로 확인하세요.
- `kwargs`는 key-value 형태이므로 `.items()`로 순회하세요.
- 짝수 판별은 `% 2 == 0`으로 할 수 있습니다.
- key가 특정 문자로 시작하는지는 문자열의 첫 글자 또는 `startswith()`로 확인할 수 있습니다.
- args에서 구한 합과 kwargs에서 구한 합을 마지막에 더해보세요.

```python
def function5(*args, **kwargs):
  result = 0
  for i in args:
    if i % 2 == 0:
      result += i
  for key, value in kwargs.items():
    if key.startswith("a"):
      result += value
  return result
```

```python
# =========================
# function1 테스트
# =========================
def test_function1():
    results = [
        function1(10, 60, 70, 50),
        function1(60, 70, 80),
        function1(10, 20, 30)
    ]
    expected = [130, 210, 0]

    for i in range(len(results)):
        if results[i] != expected[i]:
            print(f"❌ function1 Test {i+1} 실패")
            print("결과:", results[i], "정답:", expected[i])
            return
    print("✅ function1 통과")


# =========================
# function2 테스트
# =========================
def test_function2():
    results = [
        function2("A123b45c"),
        function2("123Hello123"),
        function2("2025Happy")
    ]
    expected = ["abc", "hello", "happy"]

    for i in range(len(results)):
        if results[i] != expected[i]:
            print(f"❌ function2 Test {i+1} 실패")
            print("결과:", results[i], "정답:", expected[i])
            return
    print("✅ function2 통과")


# =========================
# function3 테스트
# =========================
def test_function3():
    results = [
        function3(a=5, bb=10, c=3),
        function3(aa=5, bb=10),
        function3(a=1, b=2)
    ]
    expected = [10, 15, 0]

    for i in range(len(results)):
        if results[i] != expected[i]:
            print(f"❌ function3 Test {i+1} 실패")
            print("결과:", results[i], "정답:", expected[i])
            return
    print("✅ function3 통과")


# =========================
# function4 테스트
# =========================
def test_function4():
    results = [
        function4([1, 2, 3, 4]),
        function4([-2, 2, 4]),
        function4([1, 3, 5])
    ]
    expected = [6, 6, 0]

    for i in range(len(results)):
        if results[i] != expected[i]:
            print(f"❌ function4 Test {i+1} 실패")
            print("결과:", results[i], "정답:", expected[i])
            return
    print("✅ function4 통과")


# =========================
# function5 테스트
# =========================
def test_function5():
    results = [
        function5(1, 2, 3, 4, a=5, b=10),
        function5(2, 4, 6, aa=3, ab=5),
        function5(1, 3, 5)
    ]
    expected = [11, 20, 0]

    for i in range(len(results)):
        if results[i] != expected[i]:
            print(f"❌ function5 Test {i+1} 실패")
            print("결과:", results[i], "정답:", expected[i])
            return
    print("✅ function5 통과")
```
