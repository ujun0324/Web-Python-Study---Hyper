##문제 1

클래스 이름 : Rectangle

설명

이 클래스는 높이 height와 밑변길이 width를 갖는 직사각형을 클래스화 한 것입니다.<br>
\_\_init\_\_, get_area, is_square, \_\_add\_\_ 메소드가 미완성입니다.<br> 이 네 메소드를 완성하세요.

클래스를 제대로 완성하면 아래와 같이 코드가 실행됩니다.

코드예시

rect1 = Rectangle(10, 20) <br>
rect2 = Rectangle(10, 10) <br>
rect3 = rect1 + rect2

print(rect1.get_area()) -> 200 <br>
print(rect1.is_square()) -> False<br>
print(rect2.get_area()) -> 100 <br>
print(rect2.is_square()) -> True <br>
print(rect3.get_area()) -> 600 <br>
print(rect3.is_square()) -> False <br>


```python
class Rectangle:
	def __init__(self, height, width):
		self.height = height
		self.width = width

	def get_area(self) -> int:
		return self.height * self.width

	def is_square(self) -> bool:
		return self.height == self.width
		"""사각형이 정사각형이면 True, 아니면 False를 return (bool type)"""
		return
	def __add__(a, b):
		width = a.width + b.width
		height = a.height + b.height
		"""width로 a.width + b.width
		height로 a.height + b.height 를 갖는 Rectangle 오브젝트를 return"""
		return Rectangle(width, height)
```

##문제 2

클래스 이름 : String

설명

이 클래스는 문자열을 클래스화 한 것입니다. 문자열은 Str 변수에 저장합니다.<br>
\_\_init\_\_, \_\_str\_\_ 메소드는 이미 완성된 상태입니다.<br>
\_\_add\_\_ 메소드가 미완성입니다. 아래와 같이 동작하도록 이 메소드를 완성하세요.<br>
- String type + String type 수행시, String 인스턴스의 Str 변수끼리 더한 String 클래스 리턴
- String type + str type 수행시, String 인스턴스의 Str 변수와 str값을 더한 String 클래스 리턴

클래스를 제대로 완성하면 아래와 같이 코드가 실행됩니다.

코드예시

s1 = String("Hello") <br>
s2 = String("World") <br>
s3 = s1 + s2 <br>
s4 = s1 + "Python" <br>

print(s1) -> Hello <br>
print(s2) -> World <br>
print(s3) -> HelloWorld <br>
print(type(s3)) -> String <br>
print(s4) -> HelloPython <br>
print(type(s4)) -> String <br>

```python
class String():
	def __init__(self, Str):
		self.Str = Str

	def __str__(self):
		return self.Str

	def __add__(a, b):
		if type(b) == String:
			return String(a.Str + b.Str)
		else:
			return String(a.Str + b)
```

##문제 3

클래스 이름 : Account

설명

계좌를 생성하고, 현금의 입출금이 가능한 클래스입니다.<br>
아래에 준비된 템플릿을 활용해서 Account 클래스를 완성하세요. <br>
\_\_init\_\_ 메소드는 이미 완성되어있습니다. <br>

클래스를 제대로 완성하면 아래와 같이 코드가 실행됩니다.

코드예시

a1, a2 = Account(50), Account(0) <br>
print(a1.balance, a2.balance, a1==a2) -> 50 0 False <br>
a1.withdraw(50) <br>
a2.withdraw(50) -> withdraw Failed <br>
print(a1.balance, a2.balance, a1==a2) -> 0 0 True <br>
a1.deposit(100) <br>
a2.deposit(50) <br>
print(a1.balance, a2.balance, a1==a2) -> 100 50 False <br>

```python
from binascii import Error
class Account:
	def __init__(self, balance):
		"""Initializer: balance가 음수일때 withdraw에 제한이 생깁니다."""
		self.balance = balance

	def deposit(self, money):
		self.balance += money
		"""현금을 balance에 입금합니다. (출력 없음)"""

	def withdraw(self, money):
		if self.balance - money >= 0:
			self.balance -= money
		else:
			print("withdraw Failed")

		"""현금을 balance에서 출금합니다 (출력 없음). balance가 양수가 아닐때 에러 메시지를 출력합니다."""

	def __eq__(a, b):
		return a.balance == b.balance
		"""두 계좌의 잔고가 같으면 True, 같지 않으면 False를 반환합니다. (bool type)"""
```

```python
import io
import contextlib

# =========================
# Rectangle 테스트
# =========================
def test_Rectangle():
    try:
        rect1 = Rectangle(10, 20)
        rect2 = Rectangle(10, 10)
        rect3 = rect1 + rect2

        results = [
            rect1.get_area(),
            rect1.is_square(),
            rect2.get_area(),
            rect2.is_square(),
            rect3.get_area(),
            rect3.is_square()
        ]
        expected = [200, False, 100, True, 600, False]

        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ Rectangle Test {i+1} 실패")
                print("결과:", results[i], "정답:", expected[i])
                return
        print("✅ Rectangle 통과")

    except AttributeError as e:
        print(f"❌ Rectangle 구현 오류 (메소드나 변수가 없음): {e}")
    except TypeError as e:
        print(f"❌ Rectangle 타입 에러 (잘못된 인자 전달 등): {e}")
    except Exception as e:
        print(f"❌ Rectangle 테스트 중 예상치 못한 에러 발생: {e}")


# =========================
# String 테스트
# =========================
def test_String():
    try:
        s1 = String("Hello")
        s2 = String("World")
        s3 = s1 + s2
        s4 = s1 + "Python"

        results = [
            s1.__str__(),
            s2.__str__(),
            s3.__str__(),
            type(s3),
            s4.__str__(),
            type(s4)
        ]
        expected = ["Hello", "World", "HelloWorld", String, "HelloPython", String]

        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ String Test {i+1} 실패")
                print("결과:", results[i], "정답:", expected[i])
                return
        print("✅ String 통과")

    except AttributeError as e:
       print(f"❌ String 구현 오류 (메소드나 변수가 없음): {e}")
    except TypeError as e:
        print(f"❌ String 타입 에러 (__add__ 등에서 잘못된 연산): {e}")
    except Exception as e:
        print(f"❌ String 테스트 중 예상치 못한 에러 발생: {e}")


# =========================
# Account 테스트
# =========================
def test_Account():
    try:
        a1, a2 = Account(50), Account(0)
        results = []

        results.append((a1.balance, a2.balance, a1 == a2))
        a1.withdraw(50)

        captured_output = io.StringIO()
        with contextlib.redirect_stdout(captured_output):
            a2.withdraw(50)

        results.append((a1.balance, a2.balance, a1 == a2))
        a1.deposit(100)
        a2.deposit(50)
        results.append((a1.balance, a2.balance, a1 == a2))

        expected = [
            (50, 0, False),     # 초기 상태
            (0, 0, True),       # 출금 후 상태
            (100, 50, False)    # 입금 후 상태
        ]

        printed_msg = captured_output.getvalue().strip()
        if printed_msg != "withdraw Failed":
            print("❌ Account Test 실패: 출금 실패 메시지가 다릅니다.")
            print("결과:", printed_msg, "정답:", "withdraw Failed")
            return

        for i in range(len(results)):
            if results[i] != expected[i]:
                print(f"❌ Account Test {i+1}단계 실패")
                print("결과:", results[i], "정답:", expected[i])
                return

        print("✅ Account 통과")

    except AttributeError as e:
        print(f"❌ Account 구현 오류 (메소드나 변수가 없음): {e}")
    except Exception as e:
        print(f"❌ Account 테스트 중 에러 발생: {e}")
```

```python
test_Rectangle()
test_String()
test_Account()
```
