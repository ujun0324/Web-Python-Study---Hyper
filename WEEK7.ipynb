### 문제 1. 중복 제거하되 순서 유지하기

리스트에서 중복된 값을 제거하되, 처음 등장한 순서는 유지해서 새 리스트를 반환하세요.

- 함수 이름: `remove_duplicates_keep_order`
- 입력값 타입: 정수들이 들어 있는 리스트
- 리턴값 타입: 중복이 제거된 정수 리스트

예시:

```python
remove_duplicates_keep_order([3, 1, 2, 3, 4, 1, 5, 2, 6])
```

출력값:

```python
[3, 1, 2, 4, 5, 6]
```

````python
def remove_duplicates_keep_order(items):
    result = []
    for i in items:
      if i in result:
        continue
      else:
        result.append(i)
    return result
````

### 문제 2. 짝수만 제거하기

정수 리스트에서 짝수를 제거하고 홀수만 담긴 새 리스트를 반환하세요.

- 함수 이름: `remove_even_numbers`
- 입력값 타입: 정수들이 들어 있는 리스트
- 리턴값 타입: 홀수만 들어 있는 정수 리스트

예시:

```python
remove_even_numbers([10, 3, 5, 8, 7, 12, 15, 20])
```

출력값:

```python
[3, 5, 7, 15]
```

````python
def remove_even_numbers(nums):
  result = []
  for i in nums:
    if i % 2 == 1:
      result.append(i)
  return result
````

### 문제 3. 인접한 숫자의 합 구하기

정수 리스트에서 서로 이웃한 두 숫자의 합을 새 리스트로 만들어 반환하세요.

- 함수 이름: `neighbor_sums`
- 입력값 타입: 정수들이 들어 있는 리스트
- 리턴값 타입: 이웃한 두 숫자의 합이 들어 있는 정수 리스트

예시:

```python
neighbor_sums([2, 4, 6, 8, 10])
```

출력값:

```python
[6, 10, 14, 18]
```

````python
def neighbor_sums(nums):
    result = []
    for i in range(len(nums) - 1):
      result.append(nums[i] + nums[i + 1])
    return result
````

### 문제 4. 2차원 리스트에서 합격자 찾기

학생 이름과 점수가 들어 있는 2차원 리스트에서 점수가 70점 이상인 학생 이름만 새 리스트에 담아 반환하세요.

- 함수 이름: `find_passed_students`
- 입력값 타입: 학생 이름과 점수가 함께 들어 있는 2차원 리스트
- 리턴값 타입: 합격한 학생 이름들이 들어 있는 문자열 리스트

예시:

```python
students = [
    ["민수", 75],
    ["지우", 92],
    ["서연", 58],
    ["현우", 81],
    ["가은", 64]
]

find_passed_students(students)
```

출력값:

```python
["민수", "지우", "현우"]
```

````python
def find_passed_students(students):
    result = []
    for name, score in students:
      if score >= 70:
        result.append(name)
    return result
````

### 문제 5. 0을 기준으로 구간 합 만들기

정수 리스트에서 `0`을 구분자로 보고, 각 구간의 합을 새 리스트에 담아 반환하세요.
`0`은 합계에 포함하지 않습니다.

- 함수 이름: `section_sums_by_zero`
- 입력값 타입: 정수들이 들어 있는 리스트
- 리턴값 타입: 구간별 합이 들어 있는 정수 리스트

예시:

```python
section_sums_by_zero([3, 1, 0, 5, -2, 7, 0, -1, 0, 4, 4])
```

출력값:

```python
[4, 10, -1, 8]
```

````python
def section_sums_by_zero(nums):
    result = []
    total = 0
    for i in nums:
      if i == 0:
        result.append(total)
        total = 0
      else:
        total += i
    result.append(total)
    return result
````

### 문제 6. class Circle   
- 이 클래스는 반지름 radius을 갖는 원을 클래스화 한 것입니다.
- \_\_init\_\_, get_area, \_\_add\_\_ 메소드가 미완성입니다. 아래 Code 및 Result를 참고해서 메소드를 완성하세요. (원주율은 3.14를 사용하세요.)
- 클래스를 제대로 완성하면 아래 [Code]를 실행했을 때 [Result]대로 출력이 됩니다.
- 함수의 파라미터를 유의하여 작성하세요.

````python
class Circle:
	def __init__(self, radius = 1):
		self.radius = radius

	def get_area(self):
		"""원의 넓이를 리턴"""
		return self.radius ** 2 * 3.14

	def __add__(a, b):
		"""Circle 클래스의 인스턴스를 반환하되,
		그 인스턴스의 radius 값은 a.radius + b.radius 와 같음"""
		return Circle(a.radius + b.radius)
````

````python
# [Code]
c1 = Circle(10)
c2 = Circle(20)
c3 = Circle()
c4 = c1 + c2
c5 = c3 + c3

print(c1.radius, c2.radius, c3.radius)
print(c4.radius, c5.radius)

print(c1.get_area())
print(c4.get_area())
print(c5.get_area())

# [Result]
"""
10 20 1
30 2
314.0
2826.0
12.56
"""
````

### 문제 7. class Message
아래의 [Code]를 실행하면 [Result]와 같은 결과가 나오도록 Message 클래스를 완성하세요  
- \_\_init\_\_클래스는 이미 완성된 상태입니다. get_reversed_message 메소드만 완성하면 됩니다.

````python
class Message:
	def __init__(self, message: str):
		self.message = message
	def get_reversed_message(self, length):
		"""message를 뒤에서 length만큼의 문자를 뒤집은 문자열 리턴
		length가 message의 길이보다 크면 message 전체를 뒤집은 문자열 리턴"""
		if len(self.message) < length :
			return self.message[::-1]
		else:
			a = self.message[:-length]
			b = self.message[-length:]
			b = b[::-1]
			return a + b
````

````python
# [Code]
message1 = Message("Hello")
message2 = Message("Spring")
print(message1.get_reversed_message(1))
print(message1.get_reversed_message(3))
print(message1.get_reversed_message(5))
print(message1.get_reversed_message(7))
print(message1.message)
print(message2.get_reversed_message(1))
print(message2.get_reversed_message(3))
print(message2.get_reversed_message(5))
print(message2.get_reversed_message(7))
print(message2.message)

# [Result]
"""
Hello
Heoll
olleH
olleH
Hello
Spring
Sprgni
Sgnirp
gnirpS
Spring
"""
````

### 문제 9. class Vocal, Guitar, Band
Vocal, Guitar, Band 클래스를 만드세요  
- Vocal 클래스, Guitar 클래스는 이미 완성된 상태이니 수정하지 마세요
- Band 클래스는 Vocal 또는 Guitar 인스턴스를 member로 갖습니다. (members에는 최소 하나 이상의 오브젝트를 입력한다고 가정.)
- 아래 [Code]를 실행했을 때 [Result]대로 출력이 되도록 Band 클래스를 완성해보세요.
- 함수의 파라미터를 유의하여 작성하세요.

````python
class Vocal:
	part = 'Vocal'
	def __init__(self, name):
		self.name = name

class Guitar:
	part = 'Guitar'
	def __init__(self, name):
		self.name = name

class Band:
	def __init__(self, *members):
		self.members = members
		"""members를 인스턴스 변수로 생성합니다. (tuple type)"""

	def get_member_names(self):
		name = []
		for i in self.members:
			name.append(i.name)
		return name
		"""members의 각 인스턴스가 가진 name 변수들을 리턴합니다. (list type)"""

	def get_member_part(self):
		part = []
		for i in self.members:
			part.append(i.part)
		return part
		"""members의 각 인스턴스의 파트들을 리턴합니다. (list type)"""
````

````python
# [Code]
P1 = Vocal("Park")
P2 = Vocal("Jo")
P3 = Guitar("Kim")
P4 = Guitar("Lee")
SKBand = Band(P1, P3, P4)
MABand = Band(P2, P4)

print(P1.name)
print(P2.name)
print(P3.name)
print(P4.name)
print(SKBand.get_member_names())
print(SKBand.get_member_part())
print(MABand.get_member_names())
print(MABand.get_member_part())

# [Result]
"""
Park
Jo
Kim
Lee
['Park', 'Kim', 'Lee']
['Vocal', 'Guitar', 'Guitar']
['Jo', 'Lee']
['Vocal', 'Guitar']
"""
````

### 문제 10. class Novel, Sci_Fi, Mystery
Novel, Sci-Fi, Mystery 클래스를 만드세요  
- Sci-Fi, Mystery 클래스는 이미 완성된 상태이니 수정하지마세요
- Novel 클래스의 \_\_init\_\_, set_price, set_rate 메소드가 미완성입니다. (Novel 클래스의 price 변수 값은 20000 그대로 놔두세요)
- 아래 [Code]를 실행했을 때 [Result]대로 출력이 되도록 Novel 클래스를 완성해보세요.

````python
class Novel:
	price = 20000
	def __init__(self, title, rate):
		self.titel = title
		self.rate = rate

	@classmethod
	def set_price(self, price):
		self.price = price

	def set_rate(self, rate):
		self.rate = rate

class Sci_Fi(Novel):
	pass

class Mystery(Novel):
	pass
````

````python
# [Code]
n1 = Novel("The Parasite", 5.0)
s1 = Sci_Fi("Project Hail Mary", 4.7)
m1 = Mystery("The Tragedy of Y", 4.9)
m2 = Mystery("Inferno", 4.5)

print(n1.price, s1.price, m1.price, m2.price)
m1.set_price(25000)
print(n1.price, s1.price, m1.price, m2.price)


print(n1.rate, s1.rate, m1.rate, m2.rate)
m1.set_rate(4.3)
print(n1.rate, s1.rate, m1.rate, m2.rate)
m2.set_rate(4.0)
print(n1.rate, s1.rate, m1.rate, m2.rate)


# [Result]
"""
20000 20000 20000 20000
20000 20000 25000 25000
5.0 4.7 4.9 4.5
5.0 4.7 4.3 4.5
5.0 4.7 4.3 4.0
"""

````

## 최종 채점 코드

코드를 실행하면 각 문제별로 `정답` 또는 `오답`이 출력됩니다.  
오답인 경우에는 내가 작성한 코드의 실행 결과와 실제 정답이 함께 출력됩니다.

마지막에는 모든 문제의 점수를 합산하여 최종 점수가 출력됩니다.

````python
import sys
from io import StringIO

# ==========================================
# 점수 설정
# ==========================================

SCORES = {
    "문제 1": 7,
    "문제 2": 7,
    "문제 3": 8,
    "문제 4": 8,
    "문제 5": 10,
    "문제 6": 10,
    "문제 7": 10,
    "문제 8": 8,
    "문제 9": 14,
    "문제 10": 18
}

total_score = 0


def check_answer(name, result, expected):
    global total_score

    score = SCORES[name]

    if result == expected:
        print(name + f": 정답 (+{score}점)")
        total_score += score
    else:
        print(name + ": 오답 (+0점)")
        print("  내 결과:", result)
        print("  정답:", expected)


# ==========================================
# 리스트 문제 테스트
# ==========================================

def test_problem_1():
    try:
        return remove_duplicates_keep_order([3, 1, 2, 3, 4, 1, 5, 2, 6])
    except Exception as e:
        return f"에러 발생: {e}"


def test_problem_2():
    try:
        return remove_even_numbers([10, 3, 5, 8, 7, 12, 15, 20])
    except Exception as e:
        return f"에러 발생: {e}"


def test_problem_3():
    try:
        return neighbor_sums([2, 4, 6, 8, 10])
    except Exception as e:
        return f"에러 발생: {e}"


def test_problem_4():
    try:
        students = [
            ["민수", 75],
            ["지우", 92],
            ["서연", 58],
            ["현우", 81],
            ["가은", 64]
        ]
        return find_passed_students(students)
    except Exception as e:
        return f"에러 발생: {e}"


def test_problem_5():
    try:
        return section_sums_by_zero([3, 1, 0, 5, -2, 7, 0, -1, 0, 4, 4])
    except Exception as e:
        return f"에러 발생: {e}"


# ==========================================
# 클래스 문제 테스트
# ==========================================

def test_circle():
    try:
        c1 = Circle(10)
        c2 = Circle(20)
        c3 = Circle()
        c4 = c1 + c2
        c5 = c3 + c3

        return [
            c1.radius, c2.radius, c3.radius,
            c4.radius, c5.radius,
            c1.get_area(), c4.get_area(), c5.get_area()
        ]
    except Exception as e:
        return f"에러 발생: {e}"


def test_message():
    try:
        message1 = Message("Hello")
        message2 = Message("Spring")

        return [
            message1.get_reversed_message(1),
            message1.get_reversed_message(3),
            message1.get_reversed_message(5),
            message1.get_reversed_message(7),
            message1.message,
            message2.get_reversed_message(1),
            message2.get_reversed_message(3),
            message2.get_reversed_message(5),
            message2.get_reversed_message(7),
            message2.message
        ]
    except Exception as e:
        return f"에러 발생: {e}"


def test_my_custom_error():
    try:
        old_stdout = sys.stdout
        sys.stdout = mystdout = StringIO()

        try:
            raise MyCustomError
        except MyCustomError as e:
            print(e)
        finally:
            sys.stdout = old_stdout

        return mystdout.getvalue().strip()

    except Exception as e:
        return f"에러 발생: {e}"


def test_band():
    try:
        P1 = Vocal("Park")
        P2 = Vocal("Jo")
        P3 = Guitar("Kim")
        P4 = Guitar("Lee")

        SKBand = Band(P1, P3, P4)
        MABand = Band(P2, P4)

        return [
            P1.name, P2.name, P3.name, P4.name,
            SKBand.get_member_names(),
            SKBand.get_member_part(),
            MABand.get_member_names(),
            MABand.get_member_part()
        ]
    except Exception as e:
        return f"에러 발생: {e}"


def test_novel():
    try:
        Mystery.price = 20000

        n1 = Novel("The Parasite", 5.0)
        s1 = Sci_Fi("Project Hail Mary", 4.7)
        m1 = Mystery("The Tragedy of Y", 4.9)
        m2 = Mystery("Inferno", 4.5)

        res1 = [n1.price, s1.price, m1.price, m2.price]

        m1.set_price(25000)
        res2 = [n1.price, s1.price, m1.price, m2.price]

        res3 = [n1.rate, s1.rate, m1.rate, m2.rate]

        m1.set_rate(4.3)
        res4 = [n1.rate, s1.rate, m1.rate, m2.rate]

        m2.set_rate(4.0)
        res5 = [n1.rate, s1.rate, m1.rate, m2.rate]

        return [res1, res2, res3, res4, res5]

    except Exception as e:
        return f"에러 발생: {e}"


# ==========================================
# 채점 실행
# ==========================================

print("=== 파이썬 스터디 7주차 실전 문제 채점 시작 ===")
print()

check_answer(
    "문제 1",
    test_problem_1(),
    [3, 1, 2, 4, 5, 6]
)

check_answer(
    "문제 2",
    test_problem_2(),
    [3, 5, 7, 15]
)

check_answer(
    "문제 3",
    test_problem_3(),
    [6, 10, 14, 18]
)

check_answer(
    "문제 4",
    test_problem_4(),
    ["민수", "지우", "현우"]
)

check_answer(
    "문제 5",
    test_problem_5(),
    [4, 10, -1, 8]
)

check_answer(
    "문제 6",
    test_circle(),
    [10, 20, 1, 30, 2, 314.0, 2826.0, 12.56]
)

check_answer(
    "문제 7",
    test_message(),
    [
        "Hello", "Heoll", "olleH", "olleH", "Hello",
        "Spring", "Sprgni", "Sgnirp", "gnirpS", "Spring"
    ]
)

check_answer(
    "문제 8",
    test_my_custom_error(),
    "MyCustomError"
)

check_answer(
    "문제 9",
    test_band(),
    [
        "Park", "Jo", "Kim", "Lee",
        ["Park", "Kim", "Lee"],
        ["Vocal", "Guitar", "Guitar"],
        ["Jo", "Lee"],
        ["Vocal", "Guitar"]
    ]
)

check_answer(
    "문제 10",
    test_novel(),
    [
        [20000, 20000, 20000, 20000],
        [20000, 20000, 25000, 25000],
        [5.0, 4.7, 4.9, 4.5],
        [5.0, 4.7, 4.3, 4.5],
        [5.0, 4.7, 4.3, 4.0]
    ]
)

print()
print("==========================================")
print(f"최종 점수: {total_score} / 100점")
print("==========================================")
````
